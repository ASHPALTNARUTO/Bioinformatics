```python
import numpy as np

class Solution:
    def custom_alignment(self, sequence_A: str, sequence_B:str, substitution: dict, gap: int ) -> tuple:
        n, m = len(sequence_A), len(sequence_B)
        D = [[0]*(m+1) for i in range(n+1)]
        max_score = 0
        max_pos = None

        for i in range(1, n+1):
            for j in range(1, m+1):
                _match = D[i-1][j-1] + substitution[sequence_A[i-1]][sequence_B[j-1]]
                _delete = D[i-1][j] + gap_penalty
                _insert = D[i][j-1] + gap_penalty
                D[i][j] = max(0, _match, _delete, _insert)
                if D[i][j] > max_score:
                    max_score = D[i][j]
                    max_pos = [(i, j)]
                elif D[i][j] == max_score:
                    max_pos.append((i, j))
        
        possible_alignments = []
        for pos in max_pos:
            i, j = pos
            alignment_A, alignment_B = "", ""
            while D[i][j] != 0:
                if D[i][j] == D[i-1][j-1] + substitution[sequence_A[i-1]][sequence_B[j-1]]:
                    alignment_A = sequence_A[i-1] + alignment_A
                    alignment_B = sequence_B[j-1] + alignment_B
                    i -= 1
                    j -= 1
                elif D[i][j] == D[i-1][j] + gap:
                    alignment_A = sequence_A[i-1] + alignment_A
                    alignment_B = "-" + alignment_B
                    i -= 1
                else:
                    alignment_A = "-" + alignment_A
                    alignment_B = sequence_B[j-1] + alignment_B
                    j -= 1

            possible_alignments.append((alignment_A, alignment_B))
        
        print(np.matrix(D))
        return possible_alignments


s=Solution()
sequence_A = "venkatankitgundala"
sequence_B = "thequickbrownfoxjumpsoverthelazydog"
gap_penalty = int(input("Gap Penalty: "))
match_score = 2
mismatch_score = -1
semi_match_score = 1
characters = set(sequence_A) | set(sequence_B)
substitution_matrix = [[0 for _ in range(len(characters) + 1)] for _ in range(len(characters) + 1)]
substitution_matrix[0][1:] = list(characters)
for i, char1 in enumerate(characters):
    substitution_matrix[i+1][0] = char1
    for j, char2 in enumerate(characters):
        if char1 == char2:
            substitution_matrix[i+1][j+1] = match_score
        elif char1 in sequence_A and char2 in sequence_A:
            substitution_matrix[i+1][j+1] = semi_match_score
        else:
            substitution_matrix[i+1][j+1] = mismatch_score

S = np.array(substitution_matrix)
print(np.matrix(S))
sub_dict = {}
header = substitution_matrix[0][1:]
for i, row in enumerate(substitution_matrix[1:]):
    char1 = row[0]
    sub_dict[char1] = {}
    for j, score in enumerate(row[1:]):
        char2 = header[j]
        sub_dict[char1][char2] = int(score)
alignments=s.custom_alignment(sequence_A,sequence_B,sub_dict,gap_penalty)
print(alignments)
```
